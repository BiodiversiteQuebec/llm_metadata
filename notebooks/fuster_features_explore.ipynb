```python
import pandas as pd
import os

print("Working Directory:", os.getcwd())

# Set paths
data_dir = "data"
excel_file = os.path.join(data_dir, "dataset_092624.xlsx")

# Verify file exists
print(f"File exists: {os.path.exists(excel_file)}")

# Read the excel file
df = pd.read_excel(excel_file)
print(f"Shape: {df.shape}")
print("Columns:")
for col in df.columns:
    print(f"- {col}")
```

```python
# Print columns as list to avoid truncation
print("Columns list:", list(df.columns))

# Check for specific columns interest
interesting_cols = ['taxons', 'valid_yn', 'reason_not_valid', 'data_type', 'geospatial_info', 'spatial_range', 'temporal_range', 'duration']
existing_cols = [c for c in interesting_cols if c in df.columns]
print(f"Found interesting columns: {existing_cols}")

# Analyze unique values for validation fields
if 'valid_yn' in df.columns:
    print("\nUnique values for valid_yn:")
    print(df['valid_yn'].unique())

if 'reason_not_valid' in df.columns:
    print("\nUnique values for reason_not_valid:")
    print(df['reason_not_valid'].unique())

if 'taxons' in df.columns:
    print("\nSample values for taxons (first 10):")
    print(df['taxons'].head(10).tolist())
```

```python
# Inspect 'species', 'reason_non_valid'
print("Sample values for 'species' (first 10):")
if 'species' in df.columns:
    print(df['species'].dropna().head(10).tolist())

print("\nUnique values for 'reason_non_valid':")
if 'reason_non_valid' in df.columns:
    print(df['reason_non_valid'].unique())

print("\nSample values for 'data_type' (first 10):")
if 'data_type' in df.columns:
    print(df['data_type'].dropna().head(10).tolist())

print("\nSample values for 'dataset_format':")
if 'dataset_format' in df.columns:
    print(df['dataset_format'].unique())
```

```python
# Detailed analysis of 'species'
print("--- Species Analysis ---")
if 'species' in df.columns:
    species_list = df['species'].dropna().unique()
    print(f"Total unique values: {len(species_list)}")
    print("Sample of 30 unique species values:")
    print(species_list[:30])

# Detailed analysis of 'reason_non_valid'
print("\n--- Reason Non Valid Analysis ---")
if 'reason_non_valid' in df.columns:
    reasons = df['reason_non_valid'].unique()
    print("Unique reasons:")
    print(reasons)
    
    # Check value counts to see frequency
    print("\nValue counts:")
    print(df['reason_non_valid'].value_counts())
```
