```python
# Groups are the same from Inaturalist API (using scientific name)

# Results will be stored as a headered csv managed with pandas
# Header: Group, Taxa count, Obs count

import pandas as pd
from pathlib import Path

GROUPS = ["Aves", "Mammalia", "Reptilia", "Amphibia", "Actinopterygii", "Insecta", "Arachnida", "Gastropoda", "Plantae", "Fungi"]

group_data_file = "notebooks/group_data.csv"

if Path(group_data_file).exists():
    group_data = pd.read_csv(group_data_file)
else:
    group_data = pd.DataFrame(columns=["group", "taxa_count", "obs_count"])

    for group in GROUPS:
        print(f"Getting data for group {group}")
        taxa_count, obs_count = get_obs_summary(group)

        print(f"... Number of taxa in group {group}: {taxa_count} with {obs_count} observations")
        
        new_data = pd.DataFrame({"group": [group], "taxa_count": [taxa_count], "obs_count": [obs_count]})
        group_data = pd.concat([group_data, new_data], ignore_index=True)

print(group_data)
```
